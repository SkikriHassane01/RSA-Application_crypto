# Étapes Principales de l'Algorithme RSA

## 1. Génération des clés :
   1. **Choisir deux grands nombres premiers distincts**, p et q.
   2. **Calculer le produit** $n = p \times q$.
   3. **Calculer la fonction d'Euler** $\phi(n) = (p-1) \times (q-1)$.
   4. **Choisir un entier e** tel que $1 < e < \phi(n)$ et que $e$ soit premier avec $\phi(n)$.
   5. **Calculer d**, l'inverse modulaire de $e$ modulo $\phi(n)$, tel que $d \times e \equiv 1 \mod \phi(n)$.

## 2. Chiffrement :
   1. **Convertir le message en un entier** $m$.
   2. **Calculer le texte chiffré** $c = m^e \mod n$.

## 3. Déchiffrement :
   1. **Calculer le message original** $m = c^d \mod n$.
